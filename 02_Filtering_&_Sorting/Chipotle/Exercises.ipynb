### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep="\t", index_col="order_id")
```

### Step 4. How many products cost more than $10.00?

```python
chipo1 = chipo.copy()

c3_function = lambda x: float(x[1:-1])

# prices = [float(value[1:-1]) for value in chipo["item_price"]]

chipo1["item_price"] = chipo1["item_price"].apply(c3_function)
products_greater10 = chipo1[chipo1["item_price"] > 10]

products_greater10
```

Step 4b. Alternative way of the step above

```python
chipo2 = chipo.copy()

# First we need to transform item_price column into the float, so it can be compared with other values
prices = [float(value[1:-1]) for value in chipo2["item_price"]]

# Reassigning the column with the converted prices
chipo2["item_price"] = prices

# Deleting the duplicates
chipo_filtered = chipo2.drop_duplicates(["item_name", "quantity", "choice_description"])

chipo_quantity1 = chipo_filtered[chipo_filtered["quantity"] == 1]

chipo_greater10 = chipo_quantity1[chipo_quantity1["item_price"] > 10]["item_name"].nunique()

chipo_greater10
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
item_price_columns = chipo_quantity1[["item_name", "item_price"]].drop_duplicates("item_name")

item_price_columns
```

### Step 6. Sort by the name of the item

```python
item_price_columns.sort_values("item_name")
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo_filtered.sort_values(by="item_price", ascending=False).head(1)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
salad = chipo_filtered[chipo_filtered["item_name"] == "Veggie Salad Bowl"]
len(salad)
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
soda = chipo_filtered[(chipo_filtered["item_name"] == "Canned Soda") & (chipo_filtered["quantity"] > 1)]

len(soda)
```
